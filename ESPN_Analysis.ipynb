```python
import pandas  as pd
```

```python
df = pd.read_csv("/content/espnfinal.csv")
df
```

Questions Asked :
1. Calculate the median number of runs scored by all players.

2. Find all players who have never been out for a duck (0 runs).

3. Identify the player with the highest batting average for each country.

4. Determine the correlation between 'Matches_Played' and 'Runs'.

5. List all players who have a career span of more than 15 years.

6. Find the player with the highest individual score in an innings.

7. Create a new column 'Run_Rate' as the average runs scored per match and find the top 5 players by this metric.

8. Calculate the average strike rate of players who started their career in the same year.

9. Find all players who have scored more than 2000 runs and have a strike rate above 50.

10. Create a new column 'Total_Boundaries' as the sum of 4s and 6s hit by each player and list the top 5 players by this metric.

```python
# 1. Calculate the median number of runs scored by all players.
median_of_runs = df["Runs"].median()
median_of_runs
```

```python
# 2. Find all players who have never been out for a duck (0 runs).
df[df["0"]==0]
```

```python
# 3. Identify the player with the highest batting average for each country.
highest_batting_average = df.loc[df.groupby("Country")["Batting_Average"].idxmax()]
highest_batting_average
```

```python
# 4. Determine the correlation between 'Matches_Played' and 'Runs'.
df["Matches_Played"].corr(df["Runs"])
```

```python
# 5. List all players who have a career span of more than 15 years.
players_above_15 = df[df["Career_Span"]>15]
print(players_above_15["Player"])
```

```python
# 6. Find the player with the highest individual score in an innings.
df.iloc[df["Highest_Inns_Score"].idxmax()]["Player"]
```

```python
# 7. Create a new column 'Run_Rate' as the average runs scored per match and find the top 5 players by this metric.
df["Run_Rate"] = df["Runs"] / df["Matches_Played"]
df.nlargest(5, "Run_Rate")["Player"]
```

```python
# 8. Calculate the average strike rate of players who started their career in the same year.
# Sample Year = 2000
players_in_2000 = df[df["Initial_Year"]==2000]
players_in_2000['Strike_Rate'].mean()
```

```python
# 9. Find all players who have scored more than 2000 runs and have a strike rate above 50.
df[(df["Runs"]>2000) & (df["Strike_Rate"]>50)]["Player"]
```

```python
# 10. Create a new column 'Total_Boundaries' as the sum of 4s and 6s hit by each player and list the top 5 players by this metric.
df["Total_Boundaries"] = df["4s"] + df["6s"]
df.nlargest(5,"Total_Boundaries")
```

```python
# CONVERSION OF FILE INTO CSV FILE
df.to_csv("/content/espnanalysed.csv")
```
