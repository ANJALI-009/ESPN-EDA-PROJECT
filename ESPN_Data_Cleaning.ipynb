```python
df = df.rename(columns={"Mat":"Matches_Played","Inns":"Innings_Played","NO":"Nots_Outs","HS":"Highest_Inns_Score","Ave":"Batting_Average","BF":"Balls_Faced","SR":"Strike_Rate"})
df
```

```python
df.isnull().any()
```

```python
df[df["Balls_Faced"].isna()]
```

```python

df['Player'] = df['Player'].fillna(0)
```

```python
df.isnull().any()
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
df["Initial_Year"] = df["Span"].str.split("-").str[0]
df["Retirement_Year"] = df["Span"].str.split("-").str[1]
df
```

```python
# The axis=1 parameter specifies that we are dropping columns (as opposed to rows, which would be indicated by axis=0)
```

```python
df = df.drop(["Span"] , axis=1)
df
```

```python
df["Highest_Inns_Score"] = df["Highest_Inns_Score"].str.split("*").str[0]
df["Balls_Faced"] = df["Balls_Faced"].str.split("+").str[0]

df
```

```python
df["Country"] = df["Player"].str.split("(").str[1]
# df
```

```python
df["Country"] = df["Country"].str.split(")").str[0]
# df
```

```python
df["Player"] = df["Player"].str.split("(").str[0].str.strip()
df
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
df["Balls_Faced"] = df["Balls_Faced"].fillna(0)

```

```python
df.isnull().any()
```

```python
df = df.astype({"Highest_Inns_Score":int,"Balls_Faced":int,"Initial_Year":int,"Retirement_Year":int})
```

```python
df[df["Balls_Faced"]=="-"]
```

```python
df[df["Player"]=="Hon.FS Jackson"] = df[df["Balls_Faced"]=="-"].replace("-",0)
```

```python
df = df.astype({"Highest_Inns_Score":int,"Balls_Faced":int,"Initial_Year":int,"Retirement_Year":int})
```

```python
df.info()
```

```python
df["Career_Span"] = df["Retirement_Year"] - df["Initial_Year"]
df
```
